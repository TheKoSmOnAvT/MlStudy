```python
sns.set(style= 'darkgrid')
data["Gender"].replace("Female", 0, inplace=True)
data["Gender"].replace("Male", 1, inplace=True)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = data["Gender"]
y = data["Height"]
z = data["Weight"]

ax.set_xlabel("Gender")
ax.set_ylabel("Height")
ax.set_zlabel("Weight")
ax.scatter(x,y,z)
plt.show()
```

```python
noGender = data.drop(columns="Gender")
```

```python
sns.scatterplot(data = noGender, x ="Height", y="Weight")
```

```python
#нормировка
stand = preprocessing.MinMaxScaler()
X = stand.fit_transform(noGender)
X = pd.DataFrame(X, index=noGender.index, columns = noGender.columns)
sns.scatterplot(data = X, x ="Height", y="Weight")
```

```python
from sklearn.cluster import KMeans
from tqdm import tqdm #отображает шкалу выполнения

countCluster = range(1,31)

dist = [KMeans(n_clusters = k, random_state = 1, n_init = 10).fit(X).inertia_ for k in tqdm(countCluster)]
rockLine = pd.DataFrame()
rockLine[r'$K$ число кластеров'] = countCluster
rockLine[r'$Q_{final}(countCluster)'] = dist
```

```python
sns.lineplot(data = rockLine, x = r'$K$ число кластеров', y = r'$Q_{final}(countCluster)', markers = True, marker = 'o').set_title("каменистая ось")
```

```python
from sklearn.cluster import KMeans
from tqdm import tqdm

countCluster = range(1,11)

dist = [KMeans(n_clusters = k, random_state = 1, n_init = 10).fit(X).inertia_ for k in tqdm(countCluster)]
rockLine = pd.DataFrame()
rockLine[r'$K$ число кластеров'] = countCluster
rockLine[r'$Q_{final}(countCluster)'] = dist
sns.lineplot(data = rockLine, x = r'$K$ число кластеров', y = r'$Q_{final}(countCluster)', markers = True, marker = 'o').set_title("каменистая ось")
```

```python
model = KMeans(n_clusters = 6, random_state = 1, n_init = 10)
model.fit(X)
noGender['Cluster'] = model.labels_
#noGender.set_index('Cluster')
```

```python
noGender.groupby('Cluster').mean()
```

```python
sns.scatterplot(data=noGender, x ="Height", y="Weight",  hue= 'Cluster')
```
